# Introduction

PESMOS is a project initiated by Earthquake Engineering Department of Indian Institute of Technology Roorkee to study the ground motion characteristics of Himalayan earthquakes using strong motion accelerographs. The project was funded by Department of Science and Technology (DST), Government of India. The strong motion accelarograph network covers various region of Indian seismic zones V, IV and some thickly populated cities of seismic zone III. Out of 298 installed accelarograms 193 station have recorded various earthquake events as shown in Figure 1.

# Parsing PESMOS Records
The pesmos datas are freely available through [website](http://pesmos.in/2011/ "PESMOS") for download. Although it is required for the users to obtain a log-in credentiol for downloading the data. The log-in credential could be optained through registration. After registration web admistator will review your application and will send confirmation mail upon successful approval of registration. After logging into website, the users can go to Downloads tab for downloading data. The data are available in two section. First section, "View/Download Ground Motion Reconds" provides events from 2005 to 2014. Second section, "View/Download Earthquake Early Warning System Data (EEWS)" provides events from 2015 to 2017. For each event ".zip" file is available containing station records. For each station three direction data, NS, EW and Z-directional are available. These files are parsed using a python script named __pesmos_reader.py__. The file is available in the [GitHub page]( Pesmos/pesmos_reader.py). This python script parse the PESMOS event records for each direction individually and returns the recorded station and event information. The information on event contains origin time, location, depth of focus, magnitude, region information and source of event data. And, information on station contains location, site class, record time, sampling rate, record duration, data direction, data PGA, data filtering info, accelarogram unit, number of data points and accelarogram. 

The record file has few issues with respect correctness of data. Some of them are listed below,

1. __Magnitude Scale :__ The magnitudes reported in the file doesn't have scale associated with it.
2. __Time Stamp : __ The time stamp in many records are incorrect.
3. __Location :__ Location of events are reported as zero.
4. __Sation Code :__ The station code used in many accelarogram is not unique. It is observed that same station code is used for different recording stations. Similarly, for a recording station differend station codes are used in different event files.
5. __File Format :__ The record file format is not unique for all events.
